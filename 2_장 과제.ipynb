2.1 #14 코드

```python
def f(x):
    y=x**2-3
    return y
a=1;b=2
TOL=10**(-4)
fa=f(a)
fb=f(b)
tot=0
p1=[]
y1=[]
while True:
    p=a+(b-a)/2
    tot+=1
    fp=f(p)
    y1.append(fp)
    p1.append(p)
    if fp==0 or (b-a)/2<TOL:
        print('n값',tot)
        print('P값',p)
        break
    if fa*fp>0:
        a=p
    else: b=p
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(1.5,1.8)
y=f(x)
plt.scatter(p1,y1)
plt.plot(x,x**2-3, c = 'g', alpha = 0.5)
```

2.1 #17 코드

```python
def f(x):
    y=x**3-x-1
    return y
a=1;b=2
TOL=10**(-4)
fa=f(a)
fb=f(b)
tot=0
p1=[]
y1=[]
while True:
    p=a+(b-a)/2
    tot+=1
    fp=f(p)
    y1.append(fp)
    p1.append(p)
    if fp==0 or (b-a)/2<TOL:
        print('횟수(Pn의 n값)',tot)
        print('p값',p)
        break
    if fa*fp>0:
        a=p
    else: b=p
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(1.2,1.6)
y=f(x)
plt.scatter(p1,y1)
plt.scatter(p1[-1],y1[-1])
plt.plot(x,y, c = 'g', alpha = 0.5)
```

```python
#'fixed-pt algorithm
import math
from random import *
def g(x):
    y=math.pi+0.5*math.sin(x/2)
    return y
a=0;b=2*math.pi
i=1
y1=[]
p2=[]
p=uniform(a,b)
N0=14
TOL=10**(-2)
while i<=N0:
    p1=g(p)
    p2.append(p)
    y1.append(p1)
    if abs(p1-p)<TOL:
        print('p값',p)
        break
    i+=1
    p=p1
print('횟수(Pn의 n값)',i)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,5)
k=np.sin(x/2)*0.5+np.pi
plt.scatter(p2,y1)
plt.scatter(p2[-1],y1[-1])
plt.plot(x,k, c = 'g', alpha = 0.5)
```
